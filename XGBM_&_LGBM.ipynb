```python
import pandas as pd                                               # datset importing
df_train=pd.read_csv('/content/Titanic_train.csv')
df_train
```

```python
import pandas as pd
df_test=pd.read_csv('/content/Titanic_test.csv')
df_test
```

```python
df = pd.concat([df_train, df_test])             # dataset concatenation
df
```

```python
gender_map={                                           #conversion of categorical to numerical
    'male':0,
    'female':1
}
df['Sex']=df['Sex'].map(gender_map)
df
```

```python
Embarked_map={                                                      #conversion of categorical to numerical
    'S':0,
    'C':1,
    'Q':2
}
df['Embarked']=df['Embarked'].map(Embarked_map)
df
```

```python
df = df.fillna(df.mean(numeric_only=True))
df['Survived'] = df['Survived'].astype(int)
df
```

```python
df=df.drop(['Cabin','Ticket'],axis=1)
df
```

```python
df=df.drop(['Name'],axis=1)
df
```

```python
x=df.drop(['Survived'],axis=1)                                     # feature extraction
x
```

```python
y=df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split                                             # train test split of the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7, random_state=23)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from lightgbm import LGBMClassifier                                              # light gradient boosting algorithm implementation
from sklearn.metrics import confusion_matrix, classification_report
```

```python
model= LGBMClassifier()                                                          # model fitting
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))               # accuracy of 88% is an excellent model to implement for model development
```

```python
import lightgbm as lgb
lgb.plot_importance(model)                               # feature importance visualization
```

```python
from xgboost import XGBClassifier                               # extra gradient boosting algorithm importing
model1=XGBClassifier()                                          # model fitting
model1.fit(x_train,y_train)
```

```python
y_pred1=model1.predict(x_test)
y_pred1
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))             # accuracy of 87% is a good model to implement for the development of the model
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score
# Define hyperparameter grid
xgb_params={
    'n_estimators':[100,200],
    'learning_rate':[0.01,0.1],
    'max_depth':[3,5],
    'subsample':[0.8,1.0]
}
# Setup cross-validation strategy
cv = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
# Grid Search
grid_xgb=GridSearchCV(estimator=model1,
                        param_grid=xgb_params,
                        cv=cv,
                        scoring='accuracy',
                        verbose=1,
                        n_jobs=-1)
# Fit
grid_xgb.fit(x_train, y_train)
# Best params & score
print("Best XGBoost Params:",grid_xgb.best_params_)
print("Best CV Accuracy:",grid_xgb.best_score_)
# Predict
y_pred_xgb = grid_xgb.predict(x_test)
print("Test Accuracy:",accuracy_score(y_test,y_pred_xgb))
```

```python
#LGBM Classifier
# Define hyperparameter grid
lgb_params={
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, -1],
    'num_leaves': [31, 50]
}
# Grid Search
grid_lgb=GridSearchCV(estimator=model,
                        param_grid=lgb_params,
                        cv=cv,
                        scoring='accuracy',
                        verbose=1,
                        n_jobs=-1)
# Fit
grid_lgb.fit(x_train, y_train)
# Best params & score
print("Best LightGBM Params:",grid_lgb.best_params_)
print("Best CV Accuracy:",grid_lgb.best_score_)
# Predict
y_pred_lgb = grid_lgb.predict(x_test)
print("Test Accuracy:",accuracy_score(y_test,y_pred_lgb))
```

```python
from sklearn.metrics import roc_curve                                    # ROC curve
fpr_lgb, tpr_lgb, _=roc_curve(y_test,y_proba_lgb)
fpr_xgb, tpr_xgb, _=roc_curve(y_test,y_proba_xgb)
plt.figure(figsize=(8, 6))
plt.plot(fpr_lgb, tpr_lgb, label='LightGBM')
plt.plot(fpr_xgb, tpr_xgb, label='XGBoost')
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#LightGBM had higher recall but slightly lower precision → better at catching positives but may trigger more false alarms.
#XGBoost had slightly higher F1 and ROC AUC → more balanced classifier.
#ROC curves show both are strong classifiers; LightGBM may converge faster.
```
