```python
import pandas as pd

# Example: assuming df is your DataFrame
#df = pd.read_csv("adult_with_headers.csv")  # or however you load your data

# Feature 1: Net Capital Gain
df['capital_net_gain'] = df['capital_gain'] - df['capital_loss']
df['capital_net_gain']
```

```python
#conclusion:-Feature1-:capital_net_gain
# Desrciption:-Net capital income (gain - loss)
```

```python
# Feature 2: Hours per Week Bucket
def bucket_hours(hours):
    if hours < 30:
        return 'part_time'
    elif hours <= 40:
        return 'full_time'
    elif hours <= 50:
        return 'overtime'
    else:
        return 'overworked'

df['hours_per_week_bucket'] = df['hours_per_week'].apply(bucket_hours)
df['hours_per_week_bucket']
```

```python
#Feature2:-hours_per_week_bucket
#Desciption:-Work hour category (part-time, full-time, etc.)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Check the skewness of numerical features
numerical_cols = ['age', 'fnlwgt','education_num', 'capital_gain',
 'capital_loss', 'hours_per_week', 'capital_net_gain']
skewness = df[numerical_cols].skew().sort_values(ascending=False)

# Display skewness values
skewness
```

```python
#Chosen Feature: capital_gain
# Reason:
#Very high positive skew (11.95)
#Most individuals have 0 gain, with few having large values
#A log transformation compresses large values and spreads small values, making the distribution more normal-like.
```

```python
# Apply log transformation to capital_gain
df['capital_gain_log'] = np.log1p(df['capital_gain'])

# Plot original vs transformed distributions
plt.figure(figsize=(12, 5))

# Original
plt.subplot(1, 2, 1)
sns.histplot(df['capital_gain'], bins=50, kde=True, color='skyblue')
plt.title('Original capital_gain')
```

```python
# Transformed
plt.subplot(1, 2, 2)
sns.histplot(df['capital_gain_log'], bins=50, kde=True, color='red')
plt.title('Log-Transformed capital_gain')

plt.tight_layout()
plt.show()
```

```python
#Reduces the impact of extreme outliers.
#Makes the distribution closer to normal,
#improving model performance for algorithms sensitive to skew (e.g., linear regression, logistic regression).
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np
```

```python
# Select numerical columns for isolation
features_for_outlier_detection = ['age', 'fnlwgt', 'education_num','capital_gain', 'capital_loss', 'hours_per_week']
# Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.02, random_state=42)
df['outlier'] = iso_forest.fit_predict(df[features_for_outlier_detection])
df['outlier']
```

```python
outlier_counts= df['outlier'].value_counts()
outlier_counts
```

```python
# Keep onl y inliers (outlier = -1 means anomaly)
df_cleaned = df[df['outlier'] == 1].copy()
df_cleaned.drop(columns=['outlier'], inplace=True)
df_cleaned
```

```python
outlier_counts, df_cleaned.shape
```

```python
#Isolation Forest Results:
#Total Outliers Detected: 652
#Total Inliers (Clean Data): 31,909
#Original Dataset Size: 32,561
#Cleaned Dataset Size: (31,909 rows, 16 columns)
```

```python
#How Outliers Affect Model Performance
#Model Bias Outliers can pull regression lines or boundaries.
#Training Instability Outliers can lead to overfitting in flexible models.
#Skewed Distributions Disrupt assumptions of normality for linear models.
#Misleading MetricsOutliers may distort evaluation metrics like RMSE or R².
```
