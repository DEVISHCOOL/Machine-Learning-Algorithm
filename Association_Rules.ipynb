```python
from mlxtend.preprocessing import  TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
te=TransactionEncoder()    # TRANSACTION ENCODER CONVERTS TRANSACTION DATA  INTO FReQUENT ITEMSET
te
```

```python
te_ary=te.fit_transform(transaction)
te_ary
```

```python
print(te_ary)
```

```python
dataset=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
dataset
```

```python
# def is a function,encode_unit is a function name
def encode_units(x):
    if x== False:
        return 0
    if x==True:
       return 1
```

```python
dataset=dataset.applymap(encode_units)
dataset
```

```python
frequent_itemsets = apriori(dataset, min_support=.10,max_len=2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=.20)
rules.head(20)
```

```python
#Last Step
frequent_itemsets = apriori(dataset, min_support=.01,max_len=2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0)
rules
```
