```python
#implementing kmeans unsupervised algorithm and evaluating
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
for i in range(2,11):
  kmeans=KMeans(n_clusters=i,random_state=23)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(2,11),wcss,marker='o')             # using elbow curve method for optimal k value
plt.title('KMeans Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (inertia)')
plt.show()
```

```python
# Apply KMeans with optimal k (e.g., 4)
kmeans = KMeans(n_clusters=4, random_state=42)
k_labels = kmeans.fit_predict(x)
k_labels
```

```python
# Add to dataframe
df['KMeans_Cluster'] = k_labels
df['KMeans_Cluster']
```

```python
label=kmeans.labels_
label
```

```python
from sklearn import metrics
print(metrics.silhouette_score(x,label))  # analyzing using silhouette score
```

```python
# it is good classification with 58% of data properly clustered and can be used for calssification
```

```python
#DBSCAN algorithm
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.8, min_samples=5)
db_labels = dbscan.fit_predict(x)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(x)
distances, indices = neighbors_fit.kneighbors(x)
distances = np.sort(distances[:, 4], axis=0)

plt.figure(figsize=(8,5))
plt.plot(distances)
plt.title("DBSCAN k-distance Graph (k=5)")
plt.xlabel("Data Points Sorted")
plt.ylabel("5th Nearest Distance")
plt.grid()
plt.show()

# Add to dataframe
df['DBSCAN_Cluster'] = db_labels
df['DBSCAN_Cluster']
```
